```python
df_ted = df_ted.drop(index = 3039)
```

```python
month = []
year = []

for date in df_ted['date']:
    month.append(date.split(' ')[0])
    year.append(date.split(' ')[1]) 
    
df_ted['month'] = month
df_ted['year'] = year

df_ted.head()
```

```python
df_ted["year"] = df_ted["year"].astype('int64')
```
